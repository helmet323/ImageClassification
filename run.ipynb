```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n-cls.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(data="cifar10", epochs=10, imgsz=32)
```

```python
# Resume training
from ultralytics import YOLO

# Load a model
model = YOLO("./runs/classify/train2/weights/last.pt")  # load a partially trained model

# Resume training
results = model.train(resume=True)
```

```python
from ultralytics import YOLO

# Load a model
model = YOLO("./runs/classify/train2/weights/best.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.predict("./images.jpeg", save=True)
```
